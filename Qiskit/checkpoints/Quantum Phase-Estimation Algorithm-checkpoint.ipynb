```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import math
import numpy as np
pi = np.pi
```

```python
"n is the number of ancilla qubits, unitary gate is chosen as S"
n = 2
circuit = QuantumCircuit(n+1,n)
```

```python
def applyH(circuit,n):
    for k in range(0,n):
        circuit.h(k)
def applyU(circuit,n):
    "Creating Controlled-S gate"
    for k in range(0,n):
        for l in range(0,pow(2,n-k-1)):
            circuit.t(k)
            circuit.t(n)
            circuit.cx(k,n)
            circuit.tdg(n)
            circuit.cx(k,n)
            circuit.barrier()
            
    circuit.barrier()
def qft_dagger(circ, n):
    """n-qubit QFTdagger on q in circ."""
    for j in range(n):
        k = (n-1) - j
        for m in range(k):
            circ.cu1(-math.pi/float(2**(k-m)), k, m)
        circ.h(k)
        circ.barrier()
```

```python
applyH(circuit,n)
circuit.x(n)
circuit.barrier()
```

```python
applyU(circuit,n)
```

```python
qft_dagger(circuit,n)
```

```python
for k in range(0,n):
    circuit.measure(k,n-k-1)
```

```python
circuit.draw(output='mpl')
```
