```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import numpy as np
pi = np.pi
```

```python
qr = QuantumRegister(4)
cr = ClassicalRegister(4)
```

```python
c = QuantumCircuit(qr,cr)
```

```python
def swap_registers(circuit,k):
    for j in range(int(np.floor(k/2.))):
        circuit.swap(j,k-j-1)
    return circuit
```

```python
def qft(circ, k):
    """n-qubit QFT on the qubits in circ."""
    for j in range(k):
        circ.h(j)
        for l in range(j+1,k):
            circ.cu1((-2*pi)/float(2**(l-j+1)),l,j)
        circ.barrier()
    swap_registers(circ,k)
```

```python
qft(c,4)
c.measure([0,1,2,3], [0,1,2,3])
```

```python
simulation = Aer.get_backend("qasm_simulator")
```

```python
results = execute(c, backend = simulation).result()
```

```python
c.draw(output='mpl')
```

```python
plot_histogram(results.get_counts())
```
